```python
import pandas as pd

mengajar = pd.read_csv('mengajar.csv', encoding='utf8')
dosen = pd.read_csv('tb_dosen.csv', encoding='utf8')
kelas = pd.read_csv('tb_kelas.csv', encoding='utf8')
krs = pd.read_csv('tb_krs.csv', encoding='utf8')
mahasiswa = pd.read_csv('tb_mahasiswa.csv', encoding='utf8')
matkul = pd.read_csv('tb_mata_kuliah.csv', encoding='utf8')
nilai = pd.read_csv('tb_nilai.csv', encoding='utf8')

mengajar.shape, dosen.shape, kelas.shape, krs.shape, mahasiswa.shape, matkul.shape, nilai.shape
```

###NO.1
Perbandingan dosen yang cenderung memberikan nilai lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya

```python
nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1)
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)
nilai_dosen = nilai_dosen.set_index(['NIDN'])
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

```python
A = pd.merge(mengajar, dosen, on='NIDN', how='left')
A.loc[:,['NIDN','nama','kode_MK']]
```

#N0, 3
Perbedaan performa akademik antara mahasiswa yang berasal dari Madrasah/SMK dan SMU

```python
nilai_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_mahasiswa = nilai_mahasiswa.rename({'nilai': 'mean_nilai mahasiswa'}, axis=1)
nilai_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
nilai_mahasiswa = nilai_mahasiswa[['nim', 'nama', 'Gender', 'mean_nilai mahasiswa']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)
nilai_mahasiswa = nilai_mahasiswa.set_index(['nim'])
nilai_mahasiswa = pd.concat([nilai_mahasiswa, sb], axis=1)
nilai_mahasiswa
```

```python
perform_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
perform_mahasiswa = perform_mahasiswa.rename({'nilai': 'mean_nilai mahasiswa'}, axis=1)
perform_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
perform_mahasiswa = perform_mahasiswa[['nim', 'Asal_Sekolah','mean_nilai mahasiswa']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)
perform_mahasiswa = perform_mahasiswa.set_index(['nim'])
perform_mahasiswa = pd.concat([perform_mahasiswa, sb], axis=1)
perform_mahasiswa
```

```python
perbedaan = perform_mahasiswa.groupby(perform_mahasiswa['Asal_Sekolah']).mean()[['mean_nilai mahasiswa']]
s_b = perform_mahasiswa.groupby(perform_mahasiswa['Asal_Sekolah']).std()[['mean_nilai mahasiswa']]
s_b = s_b.rename({'mean_nilai mahasiswa': 'std_nilai Asal_Sekolah'}, axis=1)
perbedaan = pd.concat([perbedaan, s_b], axis=1)
perbedaan
```


**Interpretasi 03**

Dilihat dari rata-rata nilai dan standar deviasi, terdapat perbedaan terhadap performa akademik antara mahasiswa yang berasal dari sekolah berbeda. #NO.6
Terdapat permasalahan atau informasi lain pada data, yaitu:


1.   ada data dan variabel yang tidak terpakai
2.   ada variabel yang berbeda namun isinya sama


```python
#Informasi tambahan yakni adanya perbedaan kemampuan kaademik berdasarkan daerah asal
nilai_mahasiswa = nilai.groupby(nilai['nim']).mean()[['nilai']]
nilai_mahasiswa = nilai_mahasiswa.rename({'nilai': 'mean_nilai mahasiswa'}, axis=1)
nilai_mahasiswa = pd.merge(mahasiswa, nilai_mahasiswa, left_on='nim', right_index=True)
nilai_mahasiswa = nilai_mahasiswa[['nim', 'nama', 'Asal_Daerah','mean_nilai mahasiswa']]
sb = nilai.groupby(nilai['nim']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1)
nilai_mahasiswa = nilai_mahasiswa.set_index(['nim'])
nilai_mahasiswa = pd.concat([nilai_mahasiswa, sb], axis=1)
nilai_mahasiswa
```

```python
md = nilai_mahasiswa.groupby(nilai_mahasiswa['Asal_Daerah']).mean()[['mean_nilai mahasiswa']]
s_b = nilai_mahasiswa.groupby(nilai_mahasiswa['Asal_Daerah']).std()[['mean_nilai mahasiswa']]
s_b = s_b.rename({'mean_nilai mahasiswa': 'std_nilai daerah'}, axis=1)
md = pd.concat([md, s_b], axis=1)
md
```
